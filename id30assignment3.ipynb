```python
import numpy as np
from sympy import Matrix

# 1️
A = np.random.randint(1, 11, (3, 3))
B = np.random.randint(1, 11, (3, 3))

print("A\n", A)
print("B\n", B)

# 2️
addition = A + B
subtraction = A - B
multiplication = np.dot(A, B)

print(" \n(A + B):\n", addition)
print(" \n(A - B):\n", subtraction)
print(" \n(A × B):\n", multiplication)

# 3️
A_sym = Matrix(A)

determinant = A_sym.det()
rank = A_sym.rank()

print("\ndetermin A", determinant)
print("Rank of A", rank)

# Inverse only exists if determinant is not zero
if determinant != 0:
    inverse = A_sym.inv()
    print("\nInverse of A\n", inverse)
else:
    print("\nA is not invertible.")
```

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Crte two vec with 15 ran flots
vector1 = np.random.rand(15)
vector2 = np.random.rand(15)

# Plot both vec
plt.figure()
plt.plot(vector1, marker='o')
plt.plot(vector2, marker='s')

plt.xlabel("Index")
plt.ylabel("Value")
plt.title("tw0 random vec")
plt.legend(["Vector 1", "Vector 2"])
plt.show()
```

```python
# a)
M1 = np.random.randint(1, 10, (4, 4))
M2 = np.random.randint(1, 10, (4, 4))

# b)
add_result = M1 + M2
sub_result = M1 - M2
mul_result = np.dot(M1, M2)

# function to plot matrix as bar chart
def plot_matrix_bar(matrix, title):
    plt.figure()
    flat = matrix.flatten()
    plt.bar(range(len(flat)), flat)
    plt.title(title)
    plt.xlabel("Element Index")
    plt.ylabel("Value")
    plt.show()

# c)
plot_matrix_bar(add_result, "Mat addi res")
plot_matrix_bar(sub_result, "Mat sub res")
plot_matrix_bar(mul_result, "mat mul res")
```
