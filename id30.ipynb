```python
# Creating a 2x2 matrix
matrix = np.array([[1, 2], [3, 4]])

# Trigonometric functions
sin_values = np.sin(matrix)
cos_values = np.cos(matrix)
tan_values = np.tan(matrix)
asin_values = np.arcsin(np.clip(matrix / np.max(matrix), -1, 1))  # Clipping for valid range [-1, 1]
acos_values = np.arccos(np.clip(matrix / np.max(matrix), -1, 1))
atan_values = np.arctan(matrix)

# Exponential and logarithm
exp_values = np.exp(matrix)
log_values = np.log(matrix)  # Natural logarithm (log base e)

# Absolute value and square root
abs_values = np.abs(matrix)
sqrt_values = np.sqrt(matrix)

# Remainder when divided by a scalar (e.g., 2)
rem_values = np.remainder(matrix, 2)

# Rounding functions
round_values = np.round(matrix)
floor_values = np.floor(matrix)
ceil_values = np.ceil(matrix)

# Display results
print("Matrix:\n", matrix)
print("\nsin:\n", sin_values)
print("\ncos:\n", cos_values)
print("\ntan:\n", tan_values)
print("\nasine:\n", asin_values)
print("\nacosine:\n", acos_values)
print("\natangent:\n", atan_values)
print("\nexp:\n", exp_values)
print("\nlog (natural):\n", log_values)
print("\nabs:\n", abs_values)
print("\nsqrt:\n", sqrt_values)
print("\nrem (remainder when divided by 2):\n", rem_values)
print("\nround:\n", round_values)
print("\nfloor:\n", floor_values)
print("\nceil:\n", ceil_values)
```

```python
import numpy as np

# Creating a 3x2 matrix with custom values
matrix_3x2 = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

# Trigonometric functions
sin_values = np.sin(matrix_3x2)
cos_values = np.cos(matrix_3x2)
tan_values = np.tan(matrix_3x2)

# For inverse trig functions, scale values to [-1, 1]
scaled_matrix = np.clip(matrix_3x2 / np.max(matrix_3x2), -1, 1)
asin_values = np.arcsin(scaled_matrix)
acos_values = np.arccos(scaled_matrix)
atan_values = np.arctan(matrix_3x2)

# Exponential and logarithmic functions
exp_values = np.exp(matrix_3x2)
log_values = np.log(matrix_3x2)  # natural log

# Absolute value and square root
abs_values = np.abs(matrix_3x2)
sqrt_values = np.sqrt(matrix_3x2)

# Remainder when divided by 2
rem_values = np.remainder(matrix_3x2, 2)

# Rounding functions
round_values = np.round(matrix_3x2)
floor_values = np.floor(matrix_3x2)
ceil_values = np.ceil(matrix_3x2)

# Display results
print("3x2 Matrix:\n", matrix_3x2)

print("\nsin:\n", sin_values)
print("\ncos:\n", cos_values)
print("\ntan:\n", tan_values)

print("\nasin:\n", asin_values)
print("\nacos:\n", acos_values)
print("\natan:\n", atan_values)

print("\nexp:\n", exp_values)
print("\nlog (natural):\n", log_values)

print("\nabs:\n", abs_values)
print("\nsqrt:\n", sqrt_values)

print("\nremainder (mod 2):\n", rem_values)

print("\nround:\n", round_values)
print("\nfloor:\n", floor_values)
print("\nceil:\n", ceil_values)
```

```python
import numpy as np

# Create a 2x3 matrix with float values
matrix = np.array([
    [2.1, 4.7, 1.3],
    [5.9, 0.8, 3.6]
])

print("Matrix:\n", matrix)

max_value = np.max(matrix)
min_value = np.min(matrix)

max_index = np.unravel_index(np.argmax(matrix), matrix.shape)
min_index = np.unravel_index(np.argmin(matrix), matrix.shape)

print("\nMax value:", max_value, "at index", max_index)
print("Min value:", min_value, "at index", min_index)


vector_length = np.linalg.norm(matrix)
print("\nVector length (Euclidean norm):", vector_length)


rowwise_sorted_desc = np.sort(matrix, axis=1)[:, ::-1]

# Entire matrix descending
entire_sorted_desc = np.sort(matrix.flatten())[::-1]

print("\nRow-wise sorted (descending):\n", rowwise_sorted_desc)
print("\nEntire matrix sorted (descending):\n", entire_sorted_desc)


sum_values = np.sum(matrix)
product_values = np.prod(matrix)

print("\nSum of all elements:", sum_values)
print("Product of all elements:", product_values)


median_value = np.median(matrix)
mean_value = np.mean(matrix)
std_value = np.std(matrix)

print("\nMedian:", median_value)
print("Mean:", mean_value)
print("Standard Deviation:", std_value)
```
